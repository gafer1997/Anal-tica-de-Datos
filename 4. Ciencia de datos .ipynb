```python
import pandas as pd
import numpy as np

# Paso 1: Generar datos simulados
# Creamos un DataFrame con datos simulados
np.random.seed(0)  # Para reproducibilidad
n = 100  # Número de filas de datos simulados
data = pd.DataFrame({
    'aerolínea': np.random.choice(['Aerolínea 1', 'Aerolínea 2', 'Aerolínea 3'], n),
    'precio': np.random.normal(loc=500, scale=100, size=n),  # Precios simulados con distribución normal
    'días_antes': np.random.randint(1, 30, size=n),  # Días antes del vuelo simulados
    'escalas': np.random.randint(0, 5, size=n),  # Número de escalas simuladas
    'clase': np.random.choice(['económica', 'ejecutiva'], n)  # Clase del boleto simulada
})

# Paso 2: Visualizar datos simulados
print("Datos simulados:")
print(data.head())

# Paso 3: Análisis de datos simulados
# Ejemplo de cómo calcular la diferencia de precio entre clases
precio_economica = data[data['clase'] == 'económica']['precio'].mean()
precio_ejecutiva = data[data['clase'] == 'ejecutiva']['precio'].mean()
diferencia_precio_clases = precio_ejecutiva - precio_economica
print("Diferencia de precio entre clases:")
print(diferencia_precio_clases)

# Paso 4: Visualización de datos simulados (opcional)
import matplotlib.pyplot as plt

# Ejemplo de visualización: Histograma de precios por aerolínea
plt.figure(figsize=(10, 6))
for aerolinea in data['aerolínea'].unique():
    plt.hist(data[data['aerolínea'] == aerolinea]['precio'], alpha=0.5, label=aerolinea)
plt.xlabel('Precio')
plt.ylabel('Frecuencia')
plt.title('Histograma de precios por aerolínea')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos aleatorios
np.random.seed(0)  # Para reproducibilidad
num_vuelos = 100  # Número total de vuelos
num_escala_max = 5  # Número máximo de escalas

# Generar precios aleatorios para los vuelos
precios = np.random.randint(100, 1000, num_vuelos)

# Generar números de escalas aleatorios para los vuelos
num_escala = np.random.randint(0, num_escala_max + 1, num_vuelos)

# Calcular el precio promedio por número de escalas
precio_promedio_por_escala = []
for i in range(num_escala_max + 1):
    precio_promedio_por_escala.append(np.mean(precios[num_escala == i]))

# Graficar los resultados
plt.figure(figsize=(10, 6))
plt.plot(range(num_escala_max + 1), precio_promedio_por_escala, marker='o')
plt.title('Variación de precio según el número de escalas')
plt.xlabel('Número de escalas')
plt.ylabel('Precio promedio')
plt.grid(True)
plt.xticks(range(num_escala_max + 1))
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos aleatorios para los precios de los vuelos
np.random.seed(0)  # Para reproducibilidad
num_vuelos = 1000  # Número total de vuelos
precio_medio = 500  # Precio medio de los vuelos
desviacion_estandar = 100  # Desviación estándar de los precios

# Precios si se compra un día antes del vuelo
precios_dia_anterior = np.random.normal(precio_medio, desviacion_estandar, num_vuelos)

# Precios si se compra varios días antes del vuelo
descuento_por_compra_previa = 50
precios_varios_dias_antes = np.random.normal(precio_medio - descuento_por_compra_previa, desviacion_estandar, num_vuelos)

# Graficar los resultados
plt.figure(figsize=(10, 6))
plt.hist(precios_dia_anterior, bins=30, alpha=0.5, label='Compra un día antes')
plt.hist(precios_varios_dias_antes, bins=30, alpha=0.5, label='Compra varios días antes')
plt.title('Diferencia en precio si se compra un día antes vs varios días antes del vuelo')
plt.xlabel('Precio')
plt.ylabel('Frecuencia')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd

# Datos simulados
data = {
    "aerolínea": ["Aerolínea 1", "Aerolínea 2", "Aerolínea 1", "Aerolínea 2", "Aerolínea 2"],
    "precio": [431.341052, 501.487332, 462.433410, 496.177636, 536.797447],
    "días_antes": [10, 26, 19, 1, 28],
    "escalas": [2, 1, 2, 1, 4],
    "clase": ["ejecutiva", "económica", "económica", "ejecutiva", "ejecutiva"]
}

df = pd.DataFrame(data)

# Diferencia de precio promedio entre clases
precio_ejecutiva = df[df["clase"] == "ejecutiva"]["precio"].mean()
precio_economica = df[df["clase"] == "económica"]["precio"].mean()
diferencia_precio = precio_ejecutiva - precio_economica
print("Diferencia de precio entre clases:", diferencia_precio)

# Exploración de datos
print("\nExploración de datos:")
print(df.describe())  # Estadísticas descriptivas
print("\nTipos de aerolíneas:")
print(df["aerolínea"].unique())  # Tipos de aerolíneas
print("\nCantidad de vuelos por aerolínea:")
print(df["aerolínea"].value_counts())  # Cantidad de vuelos por aerolínea

# Visualización
import seaborn as sns
import matplotlib.pyplot as plt

# Gráfico de barras para la cantidad de vuelos por aerolínea
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x="aerolínea")
plt.title("Cantidad de vuelos por aerolínea")
plt.xlabel("Aerolínea")
plt.ylabel("Cantidad de vuelos")
plt.show()

# Gráfico de dispersión para el precio y días antes de la salida
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x="días_antes", y="precio", hue="clase", style="aerolínea", s=100)
plt.title("Precio vs. Días antes de la salida")
plt.xlabel("Días antes de la salida")
plt.ylabel("Precio")
plt.legend(title="Clase")
plt.grid(True)
plt.show()
```
